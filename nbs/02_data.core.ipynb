## Image

```python
#export
def open_img(fn):
    if not os.path.exists(fn): raise ValueError(f'File {fn} does not exists')
    return cv2.cvtColor(cv2.imread(str(fn)), cv2.COLOR_BGR2RGB)
```

```python
fns = get_image_files('samples')
im = open_img(fns[1])
```

```python
test_fail(lambda: open_img('adsfk.jpg'), msg='File adsfk.jpg does not exists')
```

```python
#export
def show_img(im, ax=None, **kwargs):
    if ax is None: fig,ax = plt.subplots(**kwargs)
    ax.set_axis_off()
    ax.imshow(im)
    return ax
```

```python
show_img(im);
```

```python
#export
def grid(f, ims, **kwargs):
    fig,axs = plt.subplots(ncols=len(ims), **kwargs)
    for fn,ax in zip(ims,axs): f(fn, ax=ax)
```

```python
ims = fns.map(open_img)
grid(show_img, ims)
```
